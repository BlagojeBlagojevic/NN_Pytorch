```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
from numpy import random



```

```python
#SIN(2PIf*t) samo ograniciti t WHITE NOICE

X=torch.arange(0,1/50,1/50000)

Y=0.3*torch.sin(100*np.pi*(X+random.rand()/1000))
Y=torch.reshape(Y,(1000,1))+random.rand()/100

A=torch.reshape(X,(1000,1))




```

```python
plt.plot(A,Y,'r--')
```

```python
from torch.nn.modules.linear import Linear


#PARAMETRI
BROJ_SEMPLOVA,BROJ_HIDEN_LAYER,EPOCH=1000,3,100

#NEURAL NETWORK KONTAM KAO DIJAMANT MADA NIJE
MODEL=nn.Sequential(
    nn.Linear(in_features=1,out_features=20,bias=True),
    
    nn.Sigmoid(),
    
    nn.Linear(in_features=20,out_features=1,bias=True),
)



```

# New Section

```python

#LOSS FUNKCJA I OPTIMAIZER
loss_funkcija=nn.L1Loss()
optimaizer=torch.optim.Adam(MODEL.parameters(),lr=0.1)
```

```python
#TRAINING

gubitci=[]
for epoch in range(20000):
  y_predi=MODEL(A)
  loss=loss_funkcija(y_predi,Y)
  gubitci.append(loss.item())
  MODEL.zero_grad()
  loss.backward()

  optimaizer.step()
```

```python
plt.plot(gubitci)
```
